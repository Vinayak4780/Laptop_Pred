```python
def fetch_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text =='Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else: 
            return 'AMD Processor'
```

```python
carpred['Cpu Brand']=carpred['Cpu Name'].apply(fetch_processor)
```

```python
carpred.drop(columns=['Cpu','Cpu Name'])
```

```python
carpred['Memory'] = carpred['Memory'].astype(str).replace('\.0', '', regex=True)
carpred["Memory"] = carpred["Memory"].str.replace('GB', '')
carpred["Memory"] = carpred["Memory"].str.replace('TB', '000')
new = carpred["Memory"].str.split("+", n = 1, expand = True)

carpred["first"]= new[0]
carpred["first"]=carpred["first"].str.strip()

carpred["second"]= new[1]

carpred["Layer1HDD"] = carpred["first"].apply(lambda x: 1 if "HDD" in x else 0)
carpred["Layer1SSD"] = carpred["first"].apply(lambda x: 1 if "SSD" in x else 0)
carpred["Layer1Hybrid"] = carpred["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
carpred["Layer1Flash_Storage"] = carpred["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

carpred['first'] = carpred['first'].str.replace(r'\D', '')

```

```python
carpred['first']=carpred['first'].str.replace(r'\D', '',regex = True)

```

```python
carpred["second"].fillna("0", inplace = True)

carpred["Layer2HDD"] = carpred["second"].apply(lambda x: 1 if "HDD" in x else 0)
carpred["Layer2SSD"] = carpred["second"].apply(lambda x: 1 if "SSD" in x else 0)
carpred["Layer2Hybrid"] = carpred["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
carpred["Layer2Flash_Storage"] = carpred["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

carpred['second'] = carpred['second'].str.replace(r'\D', '',regex = True)

carpred["first"] = carpred["first"].astype(int)
carpred["second"] = carpred["second"].astype(int)

carpred["HDD"]=(carpred["first"]*carpred["Layer1HDD"]+carpred["second"]*carpred["Layer2HDD"])
carpred["SSD"]=(carpred["first"]*carpred["Layer1SSD"]+carpred["second"]*carpred["Layer2SSD"])
carpred["Hybrid"]=(carpred["first"]*carpred["Layer1Hybrid"]+carpred["second"]*carpred["Layer2Hybrid"])
carpred["Flash_Storage"]=(carpred["first"]*carpred["Layer1Flash_Storage"]+carpred["second"]*carpred["Layer2Flash_Storage"])

carpred.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

```python
def cat_os(inp):
    if inp == 'Window 10' or inp == 'Window 7' or inp == 'Window10 5':
        return 'Window'
    else:
        if inp == 'macOS' or inp == 'Mac OS X':
            return 'Mac'
        else:
            return "Others/No os/Linux"
```

```python
x = carpred.drop(columns = ['Price'])
y = np.log(carpred['Price'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
```

```python
trans_data = ColumnTransformer(transformers = [
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')
Transfer_data = RandomForestRegressor(n_estimators = 100, random_state=3, max_samples = 0.5,max_features = 0.75,max_depth=15)
pipe = Pipeline([
    ('trans_data',trans_data),
    ('Transfer_data',Transfer_data)
])
```

```python
pipe.fit(x_train,y_train)

```

```python
y_pred = pipe.predict(x_test)
```
